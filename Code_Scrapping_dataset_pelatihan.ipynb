# Import Library

```python
from google_play_scraper import reviews
import pandas as pd
from google.colab import files
import csv
```

# Scraping

Data yang digunakan berupa review aplikasi Whatsapp dari Playstore

```python
result, _ = reviews(
    'com.whatsapp',
    lang='id',
    country='id',
    count=15000,
    filter_score_with=None
)

for r in result:
    print(r['userName'], r['score'], r['content'])
```

# Menyimpan Dataset dalam dataframe

```python
df = pd.DataFrame(result)
df.to_csv("dataset_latih_scraping.csv", index=False)
```
